```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
 
# Read the whole text.
text_file = open('./text.txt','r')
text = text_file.read()
text_file.close()
# read the mask / color image taken from
# http://jirkavinse.deviantart.com/art/quot-Real-Life-quot-Alice-282261010
alice_coloring = np.array(Image.open('./meituan.png'))
stopwords = set(STOPWORDS)
stopwords.add("said")
 
wc = WordCloud(background_color="white", max_words=2000, mask=alice_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42)
# generate word cloud
wc.generate(text)
 
# create coloring from image
image_colors = ImageColorGenerator(alice_coloring)
 
# show
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure()
wc.to_file('./meituan1.png')
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
wc.to_file('./meituan2.png')
plt.axis("off")
plt.figure()
plt.imshow(alice_coloring, cmap="gray", interpolation="bilinear")
plt.axis("off")
wc.to_file('./meituan3.png')
plt.show()
```

```python
def pie_scroll_legend() -> Pie:
    c = (
        Pie()
        .add(
            "",
            [
                list(z)
                for z in zip(
                    s.index.tolist() ,
                    s.tolist(),
                )
            ],
            center=["40%", "50%"],
            rosetype='radius'
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Pie-Legend 滚动"),
            legend_opts=opts.LegendOpts(
                type_="scroll", pos_left="80%", orient="vertical"
            ),
        )
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}.4%"))
    )
    return c
pie_scroll_legend().render_notebook()
```
