```python
import pandas as pd
#读取数据
df = pd.read_csv('./meituan1.csv')
IsDuplicated = df.duplicated()

# 去除重复店铺
data = df.drop_duplicates(['shopId'])
get = data.shape[0]

# 去除值为NaN的无用数据
data = data.dropna()

# 去除值为0的空数据
columns = data.columns.tolist()
for i in columns:
    data=data[~data[i].isin([0])] 
use = data.shape[0]
use/get
```

```python
# 数据可用地图
from pyecharts import options as opts
from pyecharts.charts import Liquid, Page
from pyecharts.globals import SymbolType

def liquid_base() -> Liquid:
    c = (
        Liquid()
        .add("lq", [0.72, 0.8])
        .set_global_opts(title_opts=opts.TitleOpts(title="可用数据占比"))
    )
    return c
liquid_base().render()
```

```python
# 商家地区分布图
area = data.groupby(['region']).size()
area.index.tolist()
area.tolist()

from example.commons import Faker
from pyecharts import options as opts
from pyecharts.charts import Map

# [list(z) for z in zip(Faker.guangdong_city, Faker.values())]
# [list(z) for z in zip(area.index.tolist(), area.tolist())]
# geo_chongqing().render_notebook()
def map_visualmap() -> Map:
    c = (
        Map()
        .add("商家A", 
             [list(z) for z in zip(area.index.tolist(), area.tolist())],
             "重庆")
        .set_global_opts(
            title_opts=opts.TitleOpts(title="商铺地理分布图"),
            visualmap_opts=opts.VisualMapOpts(max_=200, is_piecewise=True),
        )
    )
    return c
map_visualmap().render()
```

```python
# 菜品平均消费柱状图
cook_type = data.groupby(['secondCatName']).mean()
cook_avePrice= cook_type['avgPrice']
cook_avePrice

import json
import os

from pyecharts import options as opts
from pyecharts.charts import Page, PictorialBar
from pyecharts.globals import SymbolType

# with open(os.path.join("fixtures", "symbol.json"), "r", encoding="utf-8") as f:
#     symbols = json.load(f)

def pictorialbar_base() -> PictorialBar:
    c = (
        PictorialBar()
        .add_xaxis(cook_avePrice.index.tolist())
        .add_yaxis(
            "",
            cook_avePrice.tolist(),
            label_opts=opts.LabelOpts(is_show=False),
            symbol_size=10,
            symbol_repeat="fixed",
            symbol_offset=[0, 0],
            is_symbol_clip=True,
            symbol=SymbolType.ROUND_RECT,
        )
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts(title="各菜系平均消费"),
            xaxis_opts=opts.AxisOpts(is_show=True),
            yaxis_opts=opts.AxisOpts(
                axislabel_opts = opts.LabelOpts(font_size=8,rotate=30),
                
                axistick_opts=opts.AxisTickOpts(is_show=True),
                axisline_opts=opts.AxisLineOpts(
                    linestyle_opts=opts.LineStyleOpts(opacity=0)
                ),
            ),
        )
    )
    return c
pictorialbar_base().render()
```

```python
import re
user_Con = data['userComment']
comment = user_Con.str.strip()
c = comment.str.split("-")
def f(x):
    tag_list= []
    for s in x:
        num = re.findall(r"\d+\.?\d*",s)
        num = ''.join(num)
        s = s.replace(num,"")
        if num == '':
            num = '0'
        tag_list += [(s+" ")*int(num)]
    return tag_list
c =c.dropna()
tag =c.map(f) 
tag[0][0].split(' ')
file = open('./text.txt','w')
def f1(x):
    for string in x:
        s_list = string.split(' ')
        for s in s_list:
            file.write(s) 
            file.write(' ')
tag.map(f1)
file.close()

from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
 
# Read the whole text.
text_file = open('./text.txt','r')
text = text_file.read()
text_file.close()
# read the mask / color image taken from
# http://jirkavinse.deviantart.com/art/quot-Real-Life-quot-Alice-282261010
alice_coloring = np.array(Image.open('./meituan.png'))
stopwords = set(STOPWORDS)
stopwords.add("said")
 
wc = WordCloud(background_color="white", max_words=2000, mask=alice_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42)
# generate word cloud
wc.generate(text)
 
# create coloring from image
image_colors = ImageColorGenerator(alice_coloring)
 
# show
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure()
wc.to_file('./meituan1.png')
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
wc.to_file('./meituan2.png')
plt.axis("off")
plt.figure()
plt.imshow(alice_coloring, cmap="gray", interpolation="bilinear")
plt.axis("off")
wc.to_file('./meituan3.png')
plt.show()
```
